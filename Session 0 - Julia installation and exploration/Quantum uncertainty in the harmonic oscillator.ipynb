````julia
using Plots
using LaTeXStrings
````

````julia
function H(n,x)
    if n == 0 return 1 end
    if n == 1 return 2 * x end
    return 2 .* x .* H(n-1,x) .- (2 * (n-1) .* H(n-2,x))
end

function psi(n,x)
    return (1/sqrt(2^n * factorial(n) * sqrt(pi))) * exp.((-x.^2) / 2) .* collect(H(n,x))
end
````

````julia
plt = plot()
x = range(-4,4,length = 1000)
n = [0,1,2,3]
anim = Animation()

for i in n
    plot!(x, psi(i,x), label = "n = " * string(i))
    frame(anim)
end

xlabel!("x")
ylabel!(L"{\Psi}_n(x)")
display(plt)
````

### Some insights:

- Python and Julia are similar in terms of the code flow but has multiple difference in terms of syntax
- In Julia, you need to be explicit when defining lists, vectors, etc. You also have to note that when performing basic arithmentic operations with vectors, like when multiplying vectors, you need to use ```.*``` instead of just ```*``` (like in matlab)
- In Julia, similar to C/C++, you have to be careful with using ``` ` ``` and ``` " ``` because the single quote will give a character while double quotes will return a string.
- Writing functions in Julia is interesting because you can basically write an entire function in just two lines where python would usually take up multiple lines.
